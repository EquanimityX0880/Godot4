Virtual Bullet system for Godot 4

Start with generating a parabola based on two arguments: Launch angle, and initial speed.

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the launch angle and initial speed.
angle = 1
launch_angle = np.deg2rad(angle)
initial_speed = 30

# Calculate the time of flight.
time_of_flight = 2 * initial_speed * np.sin(launch_angle) / 9.8

# Calculate the range.
range = initial_speed * time_of_flight * np.cos(launch_angle)

# Create a list of time values.
t = np.linspace(0, time_of_flight, 100)

### This section is most important
# Calculate the position of the projectile at each time value.
x = initial_speed * t * np.cos(launch_angle)
y = initial_speed * t * np.sin(launch_angle) - 0.5 * 9.8 * t**2
###

# Plot the trajectory of the projectile.
plt.plot(x, y, marker = 'o', markersize=1, linestyle='')
plt.xlabel("Distance")
plt.ylabel("Height")
plt.title("Projectile Trajectory")
plt.xlim(0,range+range*1.2)
plt.ylim(0,time_of_flight/2)
plt.show()
```
